```python
import numpy as np 
import cv2 
  
# Open the image. 
img = cv2.imread('360.jpg') 
  
# Load the mask. 
mask = cv2.imread('mask1.png', 0) 
  
# Inpaint. 
dst = cv2.inpaint(img, mask, 3, cv2.INPAINT_NS) 
  
# Write the output. 
cv2.imwrite('Bike_inpainted.jpg', dst) 
```
